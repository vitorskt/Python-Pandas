```python
juncao_df = pd.DataFrame(columns=["razao_social", "nome_fantasia","nome_grupo", "desc_atividade", "logradouro", "numero", "bairro", "status"])
juncao_df['razao_social'] = df_empresas.razao_social
juncao_df['nome_fantasia'] = df_enderecos.nome_fantasia
juncao_df['nome_grupo'] = df_empresas.nome_grupo
juncao_df['desc_atividade'] = df_empresas.desc_atividade
juncao_df['logradouro'] = df_empresas.desc_atividade
juncao_df['numero'] = df_enderecos.numero
juncao_df['bairro'] = df_enderecos.nome_bairro
juncao_df['status'] = df_enderecos.situacao_empresa
```
